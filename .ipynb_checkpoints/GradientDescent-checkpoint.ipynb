```python
#Just a representation of data

x = np.random.randn(10, 1)
noise = np.random.rand()  
y = 2 * x + noise
x_flat = x.flatten()
y_flat = y.flatten()


plt.figure(figsize=(8, 6))
plt.scatter(x_flat, y_flat, color='b', marker='o', label='Data Points')
plt.plot(x_flat, 2 * x_flat + noise, color='r', linestyle='--', label='Trend Line (y = 2*x + noise)')


plt.title('Scatter Plot of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()


plt.grid(True)
plt.show()
```

```python
w = 0.0 #goal is to find m and b
```

```python
b = 0.0
```

```python
learning_rate = 0.01 #hyperparamater
```

```python
#Gradient Descent Algorithm
def descend(x,y,w,b,learning_rate):
    dldw = 0.0 #partial derivative with respect to w
    dldb = 0.0 #partial derivative with respect to c
    N = x.shape[0] #number of inputs
    for xi,yi in zip(x,y):
        #loss = (y-(wx+b))**2 ,here wx+b = yhat
        yhat = w * xi + b
        dldw += -2 * xi * (yi - yhat)
        dldb += 2 * (yi - yhat)
    w = w -learning_rate*(1/N)*dldw
    b = b -learning_rate*(1/N)*dldb
    return w,b
```

```python
for epoch in range(400):
    w,b = descend(x,y,w,b,learning_rate)
    yhat = w*x+b
    loss = np.mean((y - yhat) ** 2) 
    
    print(f'Epoch {epoch}, Loss: {loss:.4f}, Parameters: w = {w[0]:.4f}, b = {b[0]:.4f}')

    
```
