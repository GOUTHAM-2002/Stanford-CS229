Gradient Descent for Linear Regression

```python

import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(10, 1)
y = 2 * x + np.random.randn(10, 1)

m = 0.0  # Slope parameter
c = 0.0  # Intercept parameter
learning_rate = 0.01  # Hyperparameter

# Gradient Descent Algorithm
def descend(x, y, m, c, learning_rate):
    dm = 0.0  # Partial derivative with respect to m
    dc = 0.0  # Partial derivative with respect to c
    N = x.shape[0]  # Number of inputs

    # Vectorized gradient calculation
    yhat = m * x + c
    dm = -2 * np.sum(x * (y - yhat))
    dc = -2 * np.sum(y - yhat)

    m = m - learning_rate * dm
    c = c - learning_rate * dc

    return m, c

for epoch in range(400):
    m, c = descend(x, y, m, c, learning_rate)
    yhat = m * x + c
    loss = np.mean((y - yhat) ** 2)
    print(f' Loss: {loss:.4f}, Parameters: m = {m:.4f}, c = {c:.4f}')

# Print the final results
print(f'Final Loss: {loss:.4f}, Parameters: m = {m:.4f}, c = {c:.4f}')

# Plot the results
plt.scatter(x, y)
plt.plot(x, yhat, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()

```
