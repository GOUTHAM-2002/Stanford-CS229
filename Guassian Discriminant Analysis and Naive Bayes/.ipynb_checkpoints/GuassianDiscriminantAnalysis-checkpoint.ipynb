```python
class GaussianDiscriminantAnalysis:
    def __init__(self):
        self.means = None
        self.covariance = None
        self.priors = None
        self.classes = None

    def fit(self, X, y):
        self.classes = np.unique(y)
        self.means = np.zeros((len(self.classes), X.shape[1]))
        self.covariance = np.zeros((X.shape[1], X.shape[1]))
        self.priors = np.zeros(len(self.classes))
        
        for i, c in enumerate(self.classes):
            X_c = X[y == c]
            self.means[i] = X_c.mean(axis=0)
            self.covariance += np.cov(X_c, rowvar=False) * X_c.shape[0]
            self.priors[i] = X_c.shape[0]
        
        self.covariance /= len(y)
        self.priors /= len(y)

    def predict(self, X):
        inv_cov = inv(self.covariance)
        likelihood = np.zeros((X.shape[0], len(self.classes)))
        
        for i, c in enumerate(self.classes):
            mean = self.means[i]
            prior = self.priors[i]
            diff = X - mean
            likelihood[:, i] = -0.5 * np.sum(diff @ inv_cov * diff, axis=1) \
                               - 0.5 * np.log(np.linalg.det(self.covariance)) \
                               + np.log(prior)
        
        return np.argmax(likelihood, axis=1)
```

```python
gda = GaussianDiscriminantAnalysis()
gda.fit(X_train, y_train)
y_pred = gda.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
def plot_decision_boundaries(X, y, model, title):
    h = .02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', marker='o')
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.title(title)
    plt.show()

# Reduce dimensionality to 2 for visualization
X_train_2d = X_train[:, :2]
X_test_2d = X_test[:, :2]

plot_decision_boundaries(X_train_2d, y_train, gda, 'Decision Boundaries with Gaussian Discriminant Analysis (Training Set)')
plot_decision_boundaries(X_test_2d, y_test, gda, 'Decision Boundaries with Gaussian Discriminant Analysis (Test Set)')
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from scipy.linalg import inv

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Convert to DataFrame for easier handling
df = pd.DataFrame(X, columns=feature_names)
df['species'] = y

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Implement Gaussian Discriminant Analysis (GDA)
class GaussianDiscriminantAnalysis:
    def __init__(self):
        self.means = None
        self.covariance = None
        self.priors = None
        self.classes = None

    def fit(self, X, y):
        self.classes = np.unique(y)
        self.means = np.zeros((len(self.classes), X.shape[1]))
        self.covariance = np.zeros((X.shape[1], X.shape[1]))
        self.priors = np.zeros(len(self.classes))
        
        for i, c in enumerate(self.classes):
            X_c = X[y == c]
            self.means[i] = X_c.mean(axis=0)
            self.covariance += np.cov(X_c, rowvar=False) * X_c.shape[0]
            self.priors[i] = X_c.shape[0]
        
        self.covariance /= len(y)
        self.priors /= len(y)

    def predict(self, X):
        inv_cov = inv(self.covariance)
        likelihood = np.zeros((X.shape[0], len(self.classes)))
        
        for i, c in enumerate(self.classes):
            mean = self.means[i]
            prior = self.priors[i]
            diff = X - mean
            likelihood[:, i] = -0.5 * np.sum(diff @ inv_cov * diff, axis=1) \
                               - 0.5 * np.log(np.linalg.det(self.covariance)) \
                               + np.log(prior)
        
        return np.argmax(likelihood, axis=1)

# Train and evaluate the GDA model
gda = GaussianDiscriminantAnalysis()
gda.fit(X_train, y_train)
y_pred = gda.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Plot decision boundaries
def plot_decision_boundaries(X, y, model, title):
    h = .02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', marker='o')
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.title(title)
    plt.show()

# Reduce dimensionality to 2 for visualization
X_train_2d = X_train[:, :2]
X_test_2d = X_test[:, :2]

# Note: You should also train the model on 2D data if you're going to visualize in 2D
gda_2d = GaussianDiscriminantAnalysis()
gda_2d.fit(X_train_2d, y_train)  # Fit the model with 2D data

plot_decision_boundaries(X_train_2d, y_train, gda_2d, 'Decision Boundaries with Gaussian Discriminant Analysis (Training Set)')
plot_decision_boundaries(X_test_2d, y_test, gda_2d, 'Decision Boundaries with Gaussian Discriminant Analysis (Test Set)')
```
